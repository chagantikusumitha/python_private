```python
ws = []
for i in df["BMI"]:
    if i < 18.5:
        ws.append("Underweight")
    elif (i>= 18.5) & (i<=24.9):
        ws.append("Healthy")
    elif (i>= 25.0) & (i<=29.9):
        ws.append("Overweight")
    elif i>= 30.0:
        ws.append("Obese")
```

# Types of probs

```python
df = pd.read_csv("tennis.csv")
```

```python
df
```

```python
pd.crosstab(df["outlook"],df["play"])
```

```python
df1 = pd.crosstab(df["outlook"],df["play"],margins=True,normalize=True)
df1
```

# Marginal Probabilities

All values are marginal values 
rest are joint probs

- p(no)
- p(yes)
- p(overcast)
- p(rainy)
- p(sunny)

```python
p_no = df1.no.All
p_no
```

```python
p_yes = df1.yes.All
p_yes
```

```python
p_overcast = df1.All.overcast
p_overcast
```

```python
p_rainy = df1.All.rainy
p_rainy
```

```python
p_sunny = df1.All.sunny
p_sunny
```

# Joint Probabilities

```python
p_no_oc = df1.no.overcast
p_no_oc
```

```python
p_no_r = df1.no.rainy
p_no_r
```

```python
p_no_s = df1.no.sunny
p_no_s
```

```python
p_yes_oc = df1.yes.overcast
p_yes_oc
```

```python
p_yes_r = df1.yes.rainy
p_yes_r
```

```python
p_yes_s = df1.yes.sunny
p_yes_s
```

# Conditional probabilities

```python
p_o_given_no = p_no_oc/p_no
p_o_given_no
```

```python
p_o_given_yes = p_yes_oc/p_yes
p_o_given_yes
```

```python
p_r_given_no = p_no_r/p_no
p_r_given_no
```

```python
p_r_given_yes = p_yes_r/p_yes
p_r_given_yes
```

```python
p_s_given_no = p_no_s/p_no
p_s_given_no
```

```python
p_s_given_yes = p_yes_s/p_yes
p_s_given_yes
```

```python
p_o_given_no,p_o_given_yes
```

```python
p_r_given_no,p_r_given_yes
```

```python
p_s_given_no,p_s_given_yes
```

```python
def prediction(no,yes):
    if no > yes:
        return "no"
    else:
        return "yes"
```

```python
new_play = []
for i in df["outlook"]:
    if i == "overcast":
        new_play.append(prediction(p_o_given_no,p_o_given_yes))
    elif i == "sunny":
        new_play.append(prediction(p_s_given_no,p_s_given_yes))
    elif i == "rainy":
        new_play.append(prediction(p_r_given_no,p_r_given_yes))
```

```python
df["new_play"] = new_play
```
