```python
import matplotlib.pyplot as plt 
```

```python
values1 = [5,2,4,6,3,7,8,9]
values2 = [2,7,3,4,5,1,8,2]

plt.plot(values1, marker = '*', linewidth = 5)
#alpha is brightness of line ranges frm 0 to 1
#line width is width of line by default its 2
plt.plot(values2, marker = "+",alpha = 0.5)
plt.show()
```

```python
values1 = [5,2,4,6,3,7,8,9]
values2 = [2,7,3,4,5,1,8,2]

plt.plot(values1, marker = '*', linewidth = 5)
#alpha is brightness of line ranges frm 0 to 1
#line width is width of line by default its 2
plt.plot(values2, marker = "+")
#legend used for graph description sort of
plt.legend(['Apple','Samsung'])
plt.grid(True)
plt.show()
```

# PIE CHART


```python
values = [95,8,90,107,42,186]
lab = ['Delhi','Banglore','Kolkata','Hyderabad','Chennai','Pune']
ex = [0.1,0,0,0,0,0]
plt.pie(values,labels = lab,explode = ex,autopct = '%.2f',shadow = True,counterclock = False)
#counter clock is responsible for showing the values in clock wise or anticlock wise direction by default its anticlockwise
#autopct gives the % pf each area 
plt.show()
```

# BAR CHART

```python
city = ['Delhi','Beijing','Washington','Tokyo','Moscow','Banglore','Hyderabad']
happiness_index = [62,40,70,65,85,92,86]
plt.xticks(rotation = 90)
plt.title("Happiness Indexes")
plt.xlabel('cities')
plt.ylabel('Happines Index')
col = ['b','m','r','g','y']
#colors keeps on repeating after certain given values
plt.bar(city,happiness_index,color = col)
plt.show()
```

# BOX PLOT

```python
#to observe the outliers we use boxplot
#outliers: values which are faraway from the actual values of data
```

```python
values1 = [5,2,4,6,3,7,8,9,15]
values2 = [2,7,3,4,5,1,8,2,17]
boxdata = [values1,values2]
labels = ['Age','Fare']
plt.boxplot(boxdata,labels = labels,patch_artist = True,vert = 1)
#patch artist gives color,vert = 1 its is vertical else if 0 its printed in horizontal
plt.show()
```

```python
v = [25,10,2,32,37,18,175]
label = ['marks']
plt.boxplot(v,labels = label)
plt.show()
```
