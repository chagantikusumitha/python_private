```python
from ipywidgets import interact,interact_manual
```

```python
@interact
def binns_analysis(col = num_cols):
    return pd.cut(data[col],4).value_counts(normalize= True)*100
```

```python
def boxplots(col):
    sns.boxplot(data[col])
```

```python
@interact
def boxplots(col = num_cols,
            col1 = cat_cols):
    return data.groupby(col1)[col].agg({'min','mean','median','max'})
```

```python
@interact_manual
def boxplots(col = num_cols,
            col1 = cat_cols):
    sns.boxplot(data[col1],data[col])
```
