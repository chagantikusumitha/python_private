```python
dice = pd.DataFrame(np.array(np.meshgrid(dice1,dice2)).T.reshape(-1,2))
```

```python
dice
```

```python
dice.columns = ['dice1','dice2']
```

```python
dice['dice_sum'] = dice['dice1'] + dice['dice2']
```

```python
dice
```

```python
dice.dice_sum.value_counts()
```

```python
pd.DataFrame(dice.dice_sum.value_counts()).sort_index()
```

```python
pd.DataFrame(dice.dice_sum.value_counts().sort_index()).reset_index()
```

```python
prob_dist_dice = pd.DataFrame(dice.dice_sum.value_counts().sort_index()).reset_index()
```

```python
prob_dist_dice.columns = ['observations','no.of.outcomes']
```

```python
prob_dist_dice
```

```python
prob_dist_dice['probability']=np.round(prob_dist_dice['no.of.outcomes']/36,6)
```

```python
prob_dist_dice
```

```python
prob_dist_dice['cummulatives'] = np.cumsum(prob_dist_dice['probability'])
```

```python
prob_dist_dice
```
