# Stratified Random Sampling(STS)

```python
df["cut"].value_counts()
```

```python
df[df["cut"] == "Ideal"].sample(1000)
df[df["cut"] == "Premium"].sample(1000)
df[df["cut"] == "Very Good"].sample(1000)
df[df["cut"] == "Good"].sample(1000)
df[df["cut"] == "Fair"].sample(1000)
```

```python
strs = pd.DataFrame()
for i in df["cut"].unique():
    strs = strs.append(df[df["cut"] == i].sample(1000))
```
